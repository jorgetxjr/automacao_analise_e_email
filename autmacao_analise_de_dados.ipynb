### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
import pyautogui as pag
import pandas as pd
import time
```

```python
pag.PAUSE = 0.5
link_data_base= "https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga"
pag.press("win")
pag.write("chorme")
pag.press("enter")
time.sleep(1)
pag.write(link_data_base)
pag.press("enter")
time.sleep(5)
pag.click(x=704, y=284,clicks=2)
time.sleep(2)
pag.click(x=1750, y=282)


```

```python
# cálculo de indicadores
caminho_arquivo = r"C:\Users\Jorge\Downloads\Vendas - Dez.xlsx"
tabela = pd.read_excel(caminho_arquivo)
#display(tabela)
faturamento=tabela["Valor Final"].sum()
print(faturamento)
qtd_produtos=tabela["Quantidade"].sum()
print(qtd_produtos)
```
