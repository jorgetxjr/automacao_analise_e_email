# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado


Referência do pyautogui: https://pyautogui.readthedocs.io/en/latest/quickstart.html

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
import pyautogui as pag
import pandas as pd
import time
import pyperclip
import assets.config as config
```

```python
pag.PAUSE = 0.5
link_data_base= "https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga"
pag.press("win")
pag.write("chorme")
pag.press("enter")
time.sleep(1)
pag.write(link_data_base)
pag.press("enter")
time.sleep(5)
pag.click(x=704, y=284,clicks=2)
time.sleep(2)
pag.click(x=1750, y=282)


```

```python
# cálculo de indicadores
caminho_arquivo = config.download_path+"\Vendas - Dez.xlsx"
tabela = pd.read_excel(caminho_arquivo)
#display(tabela)
faturamento=tabela["Valor Final"].sum()
print(faturamento)
qtd_produtos=tabela["Quantidade"].sum()
print(qtd_produtos)
```

### Vamos agora enviar um e-mail pelo gmail

```python
mail_app = "gmail.com"
destinatario=config.destinatario
titulo_email = "Relatório de vendas - resumido"
texto_do_email=f"""
Prezados, 
segue o relatório de vendas de ontem.

Faturamento: R${faturamento:,.2f}
Quantidade de produtos vendidos: {qtd_produtos:,}

Qualquer dúvida estou à disposição
Atenciosamente,
Jorge Teixeira II
"""
pag.press("win")
pag.write("edge") #meu email está logado no edge não no chrome
pag.press("enter")
time.sleep(3)
pag.hotkey("win","up")
pag.write(mail_app)
pag.press("enter")
time.sleep(6)
pag.click(x=76, y=177)   
time.sleep(2)
pag.write(destinatario)
pag.press("tab")
pyperclip.copy(titulo_email)
pag.hotkey("ctrl","v")
pag.press("tab")
pyperclip.copy(texto_do_email)
pag.hotkey("ctrl","v")
pag.hotkey("ctrl", "enter")
```
